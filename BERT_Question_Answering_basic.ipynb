```python
import pandas as pd
import torch
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline
```

```python
inputs = [
    {"question": "What is the capital of France?", "context": "France is a country in Europe. The capital of France is Paris."},
    {"question": "Who wrote the play 'Hamlet'?", "context": "William Shakespeare was an English playwright. He wrote many famous plays, including 'Hamlet'."},
    {"question": "When was the Declaration of Independence signed?", "context": "The United States Declaration of Independence was signed on July 4, 1776."},
]
```

```python
# Select from the hub of pretrained models
model_name = 'deepset/roberta-base-squad2'
```

```python
# Load model and tokenizer
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
inputs0 = tokenizer(inputs[0]['question'], inputs[0]['context'], return_tensors="pt")
output0 = model(**inputs0)
```

```python
inputs1 = tokenizer(inputs[1]['question'], inputs[1]['context'], return_tensors="pt")
output1 = model(**inputs1)
```

```python
output0
```

```python
answer_starting_index = torch.argmax(output0.start_logits)
answer_ending_index = torch.argmax(output0.end_logits)

answer_tokens=inputs0.input_ids[0, answer_starting_index: answer_ending_index + 1]
answer = tokenizer.decode(answer_tokens)
print("Question : {}\nAnswer : {}".format(inputs[0]['question'], answer))
```

```python
# Using pipeline
qa = pipeline('question-answering', model=model_name, tokenizer=model_name)
```

```python
output_0 = qa(inputs[0]['question'], inputs[0]['context'])
print(output_0)
```

```python
output_1 = qa(inputs[1]['question'], inputs[1]['context'])
print(output_1)
```
